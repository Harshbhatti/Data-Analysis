```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url='https://internshala.com/jobs/data-science-jobs/'
```

```python
req=requests.get(url)
```

```python
req
```

```python
content=BeautifulSoup(req.content,'html.parser')
print(content.prettify())
```

```python
name=content.find_all('div',{'class':'company'})
#print(name)
#print(type(name))
#name=name[0]
#name=name.text
len(name)
```

```python
salary=content.find_all('div',{'class':'item_body salary'})
#print(salary)
len(salary)
```

```python
experience=content.find_all('div',{'class':'item_body mobile-text'})
#print(experience)
len(experience)
```

```python
nm=[]
slr=[]
exp=[]
```

```python
for i in name:
    nm.append(i.text)
    
for i in salary:
    slr.append(i.text)
    
for i in experience:
    exp.append(i.text)  
```

```python
data={'Name':nm,'Salary':slr,'Experience':exp}
```

```python
df=pd.DataFrame(data)
```
