```python
import pandas as pd
```

```python
data=pd.read_html('https://www.espncricinfo.com/rankings/content/page/211271.html')
```

# Add Data

```python
df=pd.DataFrame()
for i in range(len(data)):
    df=df.append(data[i])
```
