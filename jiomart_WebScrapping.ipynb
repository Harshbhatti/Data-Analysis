```python
#Women footwear
wf=content.find_all('div',{'class':'plp-card-details-name line-clamp jm-body-xs jm-fc-primary-grey-80'})
```

```python
name=content.find_all('div',{'class':'plp-card-details-name line-clamp jm-body-xs jm-fc-primary-grey-80'})
#print(name)
#print(type(name))
#print(name)
#name=name[0]
#name=name.text
#print(name)
len(name)
```

```python
price=content.find_all('div',{'class':'plp-card-details-price'})
#print(price)
#price=price[0].text
len(price)
```

```python
nm=[]
pr=[]
```

```python
for i in name:
    nm.append(i.text)

for i in price:
    pr.append(i.text)
```

```python
data={'Name':nm,'Price':pr}
df=pd.DataFrame(data)
```

```python
df2=df
```

```python
df2
```

```python
ss=r'[₹,,]'
```

```python
Df2=df2.applymap(lambda X:re.sub(ss,'',X))
```

```python
Df2.head()
```

```python
pattern = r'(\d+(?:\.\d+)?)'
```

```python
Df2[['Min_Price', 'Max_Price']] = Df2['Price'].str.extractall(pattern).unstack()
```

```python
Df2.head()
```

```python
Df2.isnull().sum()
```

```python
Df2['Price']=Df2['Price'].str.replace(',', '').str.split().apply(lambda x:[float(i) for i in x])
```

```python
Df2['Avg_Price']=Df2['Price'].apply(lambda x:sum(x)/len(x))
```

```python
Df2=Df2.drop('Price',axis=1)
```

```python
Df2
```

```python
Df2[]
```

```python
df3=Df2[Df2['Avg_Price']<=1000]
```

```python
df3
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(df3['Avg_Price'],labels=df3['Name'],autopct='%1.1f%%',wedgeprops={'edgecolor':'k'})
plt.axis('equal')
plt.show()
```
