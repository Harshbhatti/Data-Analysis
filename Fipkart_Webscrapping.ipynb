```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=073c3fed-722b-43b6-b956-eec3b10f282c&as-searchtext=lap"
```

```python
req=requests.get(url) # requests.get method use to get imformation in url
```

```python
# 200 means to get information in url
# 404 means error
req
```

```python
# show data in html from but not structured manner for 
# solving this problem we have used BeautifulSoup libary
```

```python
content=BeautifulSoup(req.content,'html.parser')
print(content.prettify())
```

```python
# All laptops Name
laptops=content.find_all('div',{'class':'_4rR01T'})
```

```python
print(laptops)
```

```python
print(len(laptops))
```

```python
laptop=laptops[0] # 1st laptop information
```

```python
# Name of laptops
name=content.find_all('div',{'class':'_4rR01T'})
#print(name)
#print(type(name))
#print(name)
#name=name[0]
#name=name.text
#print(name)
len(name)
```

```python
# Rating of laptops
rating=content.find_all('div',{'class':'_3LWZlK'})
#rating=rating[0].text
#print(rating)
len(rating)
```

```python
# Price of Laptop
price=content.find_all('div',{'class':'_30jeq3 _1_WHN1'})
#print(price)
#price=price[0].text
len(price)
```

```python
nm=[]
pr=[]
rt=[]
```

```python
for i in name:
    nm.append(i.text)

for i in price:
    pr.append(i.text)
    
for i in range(len(nm)):
    rt.append(rating[i].text)
```

```python
data={'NAME':nm,'PRICE':pr,'RATING':rt}
df=pd.DataFrame(data)
```
